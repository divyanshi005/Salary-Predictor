## **feature engineering:**

```python
from sklearn.preprocessing import LabelEncoder
Label_Encoder=LabelEncoder()
```

```python
sal_data1['Gender_Encode']=Label_Encoder.fit_transform(sal_data1['Gender'])
sal_data1
```

```python
sal_data1['Degree_Encode']=Label_Encoder.fit_transform(sal_data1['Degree'])
sal_data1
```

```python
sal_data1['Job_Tittle_Encode']=Label_Encoder.fit_transform(sal_data1['Job_Title'])
sal_data1
```

##Feature Scaling

```python
from sklearn.preprocessing import StandardScaler
std_scaler=StandardScaler() #object
```

```python
sal_data1['Age_scaled']=std_scaler.fit_transform(sal_data1[['Age']])
sal_data1['Experience_years_scaled']=std_scaler.fit_transform(sal_data1[['Experience_years']])
#sal_data1['Salary_scaled']=std_scaler.fit_transform(sal_data1[['Salary']])
```

## Dependant and independant features

```python
x=sal_data1[['Age_scaled','Gender_Encode','Degree_Encode','Job_Tittle_Encode','Experience_years_scaled']]
# Use a list of column names enclosed in double square brackets to select multiple columns

y=sal_data1['Salary']
```

```python
x.head()
```


**Splitting the data into training and testing**

```python
from sklearn.model_selection import train_test_split
```

train_test_split functions returns 4 values

```python
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.2,random_state=42)
```

The random_state parameter in train_test_split controls the shuffling of the data before splitting it into training and testing sets.

If you change the value of random_state, the data will be shuffled differently, resulting in different training and testing sets. This can lead to variations in model performance, as the model will be trained and evaluated on different subsets of the data.

Using a fixed random_state ensures that you get the same split every time you run the code, which is important for reproducibility and consistent evaluation of your model.

If you don't set a random_state, a random split will be generated each time, which can be useful for exploring the impact of different data splits on your model's performance.

## Model Development

```python
from sklearn.linear_model import LinearRegression
```

```python
Linear_Regression_model=LinearRegression()
#object
#This line creates a Linear Regression model object in scikit-learn. It initializes an instance of the LinearRegression class and assigns it to the variable Linera_Regression_model.

#This object will be used to train a linear regression model on your data.
```

## Model Training

```python
Linear_Regression_model.fit(x_train,y_train)
# the output shows that it i successfully fitted to the training data
```

### Model Predictions

```python
y_pred_lr=Linear_Regression_model.predict(x_test)
y_pred_lr
```

```python
df=pd.DataFrame({'y_Actual':y_test,'y_Predicted':y_pred_lr})
df
#error will come
```

```python
df=pd.DataFrame({'y_Actual':y_test,'y_Predicted':y_pred_lr})

df['Error']=df['y_Actual']-df['y_Predicted']
df['abs_error']=abs(df['Error'])

df
```

```python
MEA=df['abs_error'].mean()
MEA
```

## Model evaluation

```python
from sklearn.metrics import accuracy_score, r2_score
from sklearn.metrics import mean_squared_error, mean_absolute_error
```

## Model accuracy

```python
r2_score(y_test,y_pred_lr)
```

## The R-squared score is a statistical measure that indicates the proportion of the variance in the dependent variable (Salary) that is predictable from the independent variables (Age, Gender, etc.).

It is a value between 0 and 1, where 1 indicates a perfect fit and 0 indicates that the model does not explain any variability in the data.




```python
print(f'Accuracy of the model: {round(r2_score(y_test,y_pred_lr),4)*100}%')
```

## Mean Absolute error

```python
var1=round(mean_absolute_error(y_test,y_pred_lr ),2)
```

```python
print(f'Mean absolute error: {var1}')
```

```python
# mean square error:
var2=round(mean_squared_error(y_test,y_pred_lr),2)
print(f'Mean square error: {var2}')
```

```python
# root mean square error:
var3=round(np.sqrt(mean_squared_error(y_test,y_pred_lr)),2)
print(f'Root mean square error: {var3}')
```
